```python
import pandas as pd
cursos = pd.read_csv('cursos-prouni.csv', sep = ',', low_memory=False)
```

```python
cursos.head(10)
```

```python
cursos.isnull().sum()
```

```python
cursos.fillna(0, inplace=True)
```

```python
cursos.isnull().sum()
```

```python
def criarInsert(linha):
    insert = "INSERT INTO `cursos_prouni` VALUE ("
    for i in linha:
        if type(i) in [float, int]:
            insert += str(i) + ","
        else:
            insert += "\"" + str(i) + "\","
    return insert[:len(insert)-1] + ")" + ";\n"
```

```python
arquivo = open('insertsSQL.sql', 'w')
```

```python
for linha in cursos.values:
    arquivo.write(criarInsert(linha))
```

```python
arquivo.close()
```
